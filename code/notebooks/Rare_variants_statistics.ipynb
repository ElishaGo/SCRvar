### count umis

```python
def get_non_ref_percent(line):
    ref_base = line['reference base'].upper()
    line.drop('reference base',inplace=True)
    mutation_cols = line[~line.index.str.startswith(('R->%s' % ref_base ,'unmutated'))].index
    mutation_umi_count = line[mutation_cols].sum()
    umi_col_no_unmutated = line[~line.index.str.startswith(('unmutated'))].index
    umi_count_no_unmutated = line[umi_col_no_unmutated].sum()
    all_umi_count = line.sum()
    percent_of_non_ref_total = round((mutation_umi_count / all_umi_count) * 100,ndigits=2)
    percent_of_non_ref_mutated = round((mutation_umi_count / umi_count_no_unmutated) * 100,ndigits=2)
    return  percent_of_non_ref_mutated, percent_of_non_ref_total

all_umi_cols = [col for col in df_m.columns if col.startswith(('R->','unmutated multiple reads','unmutated single reads'))]
df_m[['percent of non ref mutated', 'percent of non ref total']] = \
    df_m[(['reference base']+all_umi_cols)].parallel_apply(get_non_ref_percent, result_type = 'expand', axis = 1)

#reorder columns
cols = ['Chromosome', 'start', 'end', 'count of unmutated cell barcodes','count of mutated cell barcodes',
       'percent of non ref total', 'percent of non ref mutated', 'strand', 'reference base', 'R->A multi reads',
       'R->T multi reads', 'R->G multi reads', 'R->C multi reads',
       'R->A single read', 'R->T single read', 'R->G single read',
       'R->C single read', 'mixed reads', 'unmutated single reads', 'unmutated multiple reads', 'position',
       'aggregated cell barcodes']
df_m = df_m[cols]
#sort df by chromosome and position
df_m.sort_values(by=['Chromosome', 'start'], inplace = True)
```

### Get regions of positions where 'percent of non ref' is below 75 in average

```python
from itertools import groupby, cycle 

percent_to_filter = 75
col = 'percent of non ref mutated'

def groupSequence(l): 
    temp_list = iter(l)
    next(temp_list)
    groups = groupby(l, key = lambda j: j + 10 >= next(temp_list)) 

    for k, v in groups: 
        if k: 
            yield tuple(v) + (next((next(groups)[1])), )

end_position = df_agg[df_agg.rolling(10, min_periods=1).mean()[col] <= percent_to_filter]['end']
tpls = list(groupSequence(end_position))
print(*tpls[:10], sep='\n')
```
