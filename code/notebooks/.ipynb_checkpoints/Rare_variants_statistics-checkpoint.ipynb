```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
from math import ceil
import swifter
import random
pd.set_option('display.max_columns', None)

out_folder = './'
```

```python
   
def filter_by_cb_count(df, df_agg, min_mutatiion_to_filter,min_nonmutatiion_to_filter):
    """
    First groupby position, then filter the positions by provided column, and return the un grouped df.
    """
    df_agg_filt = df_agg[(df_agg['count of mutated cell barcodes'] >= min_mutatiion_to_filter) & (df_agg['count of unmutated cell barcodes'] >= min_nonmutatiion_to_filter)]
    filter_idx = df_agg_filt['position'].values
    df_filt = df[df['position'].isin(filter_idx)]
    return df_filt, df_agg_filt

def agg_dfs(df):
    """
    Aggregate the mutated dataframe on the base position.The basic aggregation function is sum.
    Also add count of CB in each position.
    """
    df_agg = df.groupby('position').agg(
        {'Chromosome': 'first', 'start': 'first', 'end': 'first', 'strand': 'first', 'reference base': 'first','R->A multi reads': 'sum', 'R->T multi reads': 'sum', 'R->G multi reads': 'sum',
         'R->C multi reads': 'sum', 'R->A single reads': 'sum', 'R->T single reads': 'sum',
         'R->G single reads': 'sum', 'R->C single reads': 'sum', 'mixed reads': 'sum',
         'cell barcode': lambda x: ','.join(x), 'count of unmutated cell barcodes': 'first', 'unmutated multi reads': 'first',
             'unmutated single reads': 'first'}).reset_index()
    df_agg.rename(columns={"cell barcode": "aggregated cell barcodes"}, inplace=True)
    df_agg.loc[:, 'count of mutated cell barcodes'] = df_agg.apply(
        lambda x: len(x['aggregated cell barcodes'].split(',')), axis=1) # add column with count of CB in each position . This can maybe be shorter by using groupby and count
    return df_agg

def merge_dfs(df_mutated, df_unmutated):
    # merge aggregated mutations and non mutations tables
    df_m = df_mutated.merge(
        df_unmutated[['position', 'count of unmutated cell barcodes', 'unmutated single reads', 'unmutated multi reads']],
        how='left', on='position')
    return df_m

def agg_unmutated(df):
    """
    Aggregate the unmutated dataframe on the base position.The basic aggregation function is sum.
    Also add count of CB in each position.
    """
    #check if aggregation is needed
    if df.duplicated('position', keep = False).sum() > 0:
        df = df.groupby('position').agg(
            {'count of unmutated cell barcodes': 'sum', 'unmutated multi reads': 'sum',
             'unmutated single reads': 'sum', 'total umi count': 'sum'}).reset_index()
    return df

def add_position(df):
    """
    Add one column with the position notation.
    """
    # concat 3 first columns into one column of position
    df['position'] = df.Chromosome.str.cat(df['start'].astype(str), sep=':')
    df['position'] = df.position.str.cat(df['end'].astype(str), sep='-')
    df['position'] = df.position.str.cat(df['strand'].astype(str), sep = ',')
    return df



df_mutated = pd.read_csv('/home/eligol/Documents/01_WIS/01_Rarevar/data/raw_Stats_files/removed/ADAR/concatenated_tsv/raw_stats_removed_ADAR.tsv', sep='\t')
df_mutated.rename(columns = {'R->A multi': 'R->A multi reads',
                     'R->T multi': 'R->T multi reads', 'R->G multi': 'R->G multi reads',
                     'R->C multi': 'R->C multi reads', 'R->A single': 'R->A single reads',
                     'R->T single': 'R->T single reads','R->G single': 'R->G single reads',
                     'R->C single': 'R->C single reads'}, inplace = True)
df_mutated = add_position(df_mutated)
mutation_cols = df_mutated.columns[df_mutated.columns.str.startswith(('R->'))]
df_mutated['total umi count'] = df_mutated[mutation_cols].sum(axis=1)

df_unmutated = pd.read_csv('/home/eligol/Documents/01_WIS/01_Rarevar/data/raw_Stats_files/removed/ADAR/concatenated_tsv/raw_stats_unmutated_removed_ADAR.tsv', sep='\t')
df_unmutated.rename(
    columns={'chromosome': 'Chromosome', 'direction': 'strand', 'unique cells': 'count of unmutated cell barcodes',
             'multiples': 'unmutated multi reads', 'singles': 'unmutated single reads'}, inplace=True)
df_unmutated['total umi count'] = df_unmutated['unmutated multi reads'] + df_unmutated['unmutated single reads']
df_unmutated = add_position(df_unmutated)
df_unmutated = agg_unmutated(df_unmutated)
    
df_merged = merge_dfs(df_mutated, df_unmutated) #NAs exisiting
df_merged_agg = agg_dfs(df_merged)

df_merged_filtered, df_merged_agg_filtered = filter_by_cb_count(df_merged, df_merged_agg, 5, 1)
```

```python
def plot_cb_occurences_hist(df_orig, df_filtered,  out_folder):
    """
    This function count the number of each Cb in the table, and then plot the distribution of CB occurences. This can give an insight regarding the representation of different cell in the table. Note that the counts in tables can overlap on positions.
    """
    fig, axs = plt.subplots(2, 1, figsize=(10, 12), sharex=True, sharey=True)

    def make_plot(ax, cb_counts, df_name):
        ax.hist(cb_counts.values, bins=50)
        ax.set_title("Number of cell barcodes - {} filtering".format(df_name))
        ax.set_ylabel("count of different cell barcodes")
        ax.set_xlabel("number of mutated positions")

    for i, df_tuple in enumerate(zip([df_orig, df_filtered], ['prior','post'])):
        df = df_tuple[0]
        df_name = df_tuple[1]
        cb_counts = df['cell barcode'].value_counts()
        make_plot(axs[i], cb_counts, df_name)

def get_stat_plots(df_merged, df_merged_agg):#(df, df_filtered, df_unmutated, df_filtered_unmutated, out_folder):
    """
    Function to manage all plots
    :param df: data frame of mutations
    :param df_filtered: data frame of mutations after filtering
    :param df_unmutated: data frame of non mutations
    :param df_filtered_unmutated: data frame of non mutations after filtering
    :param out_folder: path to folder to save outputs
    """
    df_merged_filtered, df_merged_agg_filtered = filter_by_cb_count(df_merged, df_merged_agg, 5, 1)
    print(df_merged_filtered.shape)
    print(df_merged_agg_filtered.shape)
    plot_cb_occurences_hist(df_merged, df_merged_filtered, out_folder)
    
get_stat_plots(df_merged, df_merged_agg)
```

### count umis

```python
def get_non_ref_percent(line):
    ref_base = line['reference base'].upper()
    line.drop('reference base',inplace=True)
    mutation_cols = line[~line.index.str.startswith(('R->%s' % ref_base ,'unmutated'))].index
    mutation_umi_count = line[mutation_cols].sum()
    umi_col_no_unmutated = line[~line.index.str.startswith(('unmutated'))].index
    umi_count_no_unmutated = line[umi_col_no_unmutated].sum()
    all_umi_count = line.sum()
    percent_of_non_ref_total = round((mutation_umi_count / all_umi_count) * 100,ndigits=2)
    percent_of_non_ref_mutated = round((mutation_umi_count / umi_count_no_unmutated) * 100,ndigits=2)
    return  percent_of_non_ref_mutated, percent_of_non_ref_total

all_umi_cols = [col for col in df_m.columns if col.startswith(('R->','unmutated multiple reads','unmutated single reads'))]
df_m[['percent of non ref mutated', 'percent of non ref total']] = \
    df_m[(['reference base']+all_umi_cols)].parallel_apply(get_non_ref_percent, result_type = 'expand', axis = 1)

#reorder columns
cols = ['Chromosome', 'start', 'end', 'count of unmutated cell barcodes','count of mutated cell barcodes',
       'percent of non ref total', 'percent of non ref mutated', 'strand', 'reference base', 'R->A multi reads',
       'R->T multi reads', 'R->G multi reads', 'R->C multi reads',
       'R->A single read', 'R->T single read', 'R->G single read',
       'R->C single read', 'mixed reads', 'unmutated single reads', 'unmutated multiple reads', 'position',
       'aggregated cell barcodes']
df_m = df_m[cols]
#sort df by chromosome and position
df_m.sort_values(by=['Chromosome', 'start'], inplace = True)
```

```python
from itertools import groupby, cycle 

percent_to_filter = 75
col = 'percent of non ref mutated'

def groupSequence(l): 
    temp_list = iter(l)
    next(temp_list)
    groups = groupby(l, key = lambda j: j + 10 >= next(temp_list)) 

    for k, v in groups: 
        if k: 
            yield tuple(v) + (next((next(groups)[1])), )

end_position = df_agg[df_agg.rolling(10, min_periods=1).mean()[col] <= percent_to_filter]['end']
tpls = list(groupSequence(end_position))
print(*tpls[:10], sep='\n')
```
